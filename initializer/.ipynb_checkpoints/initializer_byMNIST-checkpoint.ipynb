```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load MNIST data 

```python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('MNIST_data', validation_size=0, one_hot=True)
```

### Check MNIST image

```python
img = mnist.train.images[100]
plt.imshow(img.reshape((28, 28)), cmap='Greys_r')
```

```python
image_size = mnist.train.images.shape[1]
n_label = 10
l1_dim = 256
l2_dim = 128
l3_dim = 10

n_loop = 200
batch_size = 50
```

### define functions

```python
def training(n_loop, batch_size):
    train_acc_ = np.zeros(n_loop)
    test_acc_ = np.zeros(n_loop)
    train_cost_ = np.zeros(n_loop)
    test_cost_ = np.zeros(n_loop)
    
    sess.run(init)
    sess.run(tf.global_variables_initializer())
    for e in range(n_loop): 
        #for _ in range(int(mnist.train.images.shape[0] / batch_size)):
        batch = mnist.train.next_batch(batch_size)
        sess.run([opt], feed_dict={x: batch[0], y_: batch[1]})

        train_cost, train_accuracy = sess.run([cost, accuracy], feed_dict={x: batch[0], y_: batch[1]})
        #print('step %d, training accuracy %g' % (e, train_accuracy)) 
        test_cost, test_accuracy = sess.run([cost, accuracy], feed_dict={x: mnist.test.images, y_: mnist.test.labels})
        #print('test accuracy %g' % (test_accuracy))
        train_acc_[e] = train_accuracy
        test_acc_[e] = test_accuracy
        train_cost_[e] = train_cost
        test_cost_[e] = test_cost
        
    return train_acc_, test_acc_, train_cost_, test_cost_

def plot_result(test_acc, train_acc, test_cost, train_cost):
    plt.subplot(1, 2, 1)
    plt.plot(test_acc, label="test_acc")
    plt.plot(train_acc, label="train_acc")
    plt.xlabel('epoch')
    plt.ylabel('acc')
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(test_cost, label="test_cost")
    plt.plot(train_cost, label="train_cost")
    plt.xlabel('epoch')
    plt.ylabel('cost')
    plt.legend()
```

### base model

```python
base = tf.Graph()
with base.as_default():
    with tf.variable_scope("base"):
        # Input and output placeholder
        init = tf.global_variables_initializer()
        
        x = tf.placeholder(tf.float32, shape=[None, image_size], name='x')
        y_ = tf.placeholder(tf.float32, shape=[None, n_label], name='y_')

        # fully connected layer here with ReLU activation
        l1 = tf.layers.dense(x, l1_dim, tf.nn.relu)
        l2 = tf.layers.dense(l1, l2_dim, tf.nn.relu)
        y = tf.layers.dense(l2, l3_dim)

        # Mean of the loss using Sigmoid cross-entropy loss
        cost = tf.losses.sigmoid_cross_entropy(y_, y)

        # Adam optimizer
        opt = tf.train.AdamOptimizer(0.001).minimize(cost)

        correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

tf.reset_default_graph()
```

```python
with tf.Session(graph=base) as sess:
    train_acc, test_acc, train_cost, test_cost = training(n_loop, batch_size)
    
plot_result(test_acc, train_acc, test_cost, train_cost)
```

### using Xavier as an initializer

```python
xavier = tf.Graph()
with xavier.as_default():
    with tf.variable_scope("xavier"):
        # Input and output placeholder
        init = tf.global_variables_initializer()
        
        x = tf.placeholder(tf.float32, shape=[None, image_size], name='x')
        y_ = tf.placeholder(tf.float32, shape=[None, n_label], name='y_')

        # fully connected layer here with ReLU activation
        ## Xavier
        initializer = tf.contrib.layers.xavier_initializer(uniform=False, seed=None, dtype=tf.float32)
        ## He
        #initializer = tf.contrib.keras.initializers.he_normal()
        l1 = tf.layers.dense(x, l1_dim, tf.nn.relu, kernel_initializer = initializer)
        l2 = tf.layers.dense(l1, l2_dim, tf.nn.relu, kernel_initializer = initializer)
        y = tf.layers.dense(l2, l3_dim, kernel_initializer = initializer)

        # Mean of the loss using Sigmoid cross-entropy loss
        cost = tf.losses.sigmoid_cross_entropy(y_, y)

        # Adam optimizer
        opt = tf.train.AdamOptimizer(0.001).minimize(cost)

        correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
tf.reset_default_graph()
```

```python
with tf.Session(graph=xavier) as sess:
    train_acc, test_acc, train_cost, test_cost = training(n_loop, batch_size)
    
plot_result(test_acc, train_acc, test_cost, train_cost)
```

### using Xavier as an initializer

```python
He = tf.Graph()
with He.as_default():
    with tf.variable_scope("He"):
        # Input and output placeholder
        init = tf.global_variables_initializer()
        
        x = tf.placeholder(tf.float32, shape=[None, image_size], name='x')
        y_ = tf.placeholder(tf.float32, shape=[None, n_label], name='y_')

        # fully connected layer here with ReLU activation
        ## He
        initializer = tf.contrib.keras.initializers.he_normal()
        l1 = tf.layers.dense(x, l1_dim, tf.nn.relu, kernel_initializer = initializer)
        l2 = tf.layers.dense(l1, l2_dim, tf.nn.relu, kernel_initializer = initializer)
        y = tf.layers.dense(l2, l3_dim, kernel_initializer = initializer)

        # Mean of the loss using Sigmoid cross-entropy loss
        cost = tf.losses.sigmoid_cross_entropy(y_, y)

        # Adam optimizer
        opt = tf.train.AdamOptimizer(0.001).minimize(cost)

        correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
        accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    
tf.reset_default_graph()
```

```python
with tf.Session(graph=He) as sess:
    train_acc, test_acc, train_cost, test_cost = training(n_loop, batch_size)
    
plot_result(test_acc, train_acc, test_cost, train_cost)
```
